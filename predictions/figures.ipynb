```python
import os
import pandas as pd
import pickle
import numpy as np
import dissertation.tools.utils as utils
import librosa
import matplotlib.pyplot as plt
import matplotlib as mpl
import subprocess
import glob
from keras.models import model_from_json
from speech2ult.tools.transform_ultrasound import transform_ultrasound

def load_model(path):
    model_weights = path[:-10] + 'weights.h5'
    with open(path, "r") as json_file:
        loaded_model_json = json_file.read()
    model = model_from_json(loaded_model_json)
    # load weight into model
    model.load_weights(model_weights)
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

class TestFile:
    def __init__(self, basename, ult, wav, sr, dlc_ult):
        self.basename = basename
        self.ult = ult
        self.wav = wav
        self.sr = sr
        self.dlc = dlc_ult
        self.aud_scalar = None
        self.ult_scalar = None
        self.dlc_scalar = None
        self.aud_feat = None
        self.window = None
        self.lstm = None
        self.get_butter()
        self.get_med()
        self.get_bothfilt()

    def get_butter(self):
        """apply butterworth filter to each column (body part)"""
        filtered = np.empty_like(self.dlc)
        fc = 10  # Cut-off frequency of the filter
        w = fc / (60 / 2)  # Normalize the frequency
        b, a = signal.butter(5, w, 'low')

        for col in range(self.dlc.shape[1]):
            filtered[:, col] = signal.filtfilt(b, a, self.dlc[:, col])

        self.butter = filtered

    def get_med(self):
        """apply median filter to each column (body part)"""
        filtered = np.empty_like(self.dlc)
        for col in range(self.dlc.shape[1]):
            filtered[:, col] = signal.medfilt(self.dlc[:, col], 7)

        self.medfilt = filtered

    def get_bothfilt(self):
        """apply median filter to each column (body part)"""
        filtered = np.empty_like(self.dlc)
        for col in range(self.dlc.shape[1]):
            filtered[:, col] = signal.medfilt(self.butter[:, col], 7)

        self.bothfilt = filtered

def transfrom(ult, colour=0):
    ult_t = transform_ultrasound(ult, background_colour=colour, num_scanlines=64,
                             size_scanline=128, angle=0.025,
                             zero_offset=30, pixels_per_mm=0.5)

    # get rid of border
    ult_t = ult_t[:, 50:-50, 10:-10]
    return ult_t

def make_vid(img_dir, name, type='avi', out_dir=None):
    if not out_dir:
        out_dir = img_dir
    subprocess.call(
        ["ffmpeg", "-y", "-r", "60",
         "-i", img_dir + "/%07d.jpg", "-vcodec", "mpeg4", "-qscale", "5", "-r",
         "60", out_dir + f"/{name}.{type}"])
```

```python
testfiles = glob.glob('speech2ult/predictions/test_data/*testfile*')
testfile_path = testfiles[1]
# load in testfile
with open(testfile_path, 'rb') as file:
    testfile = pickle.load(file)

# fix testfile.transform
testfile.transform = transfrom(testfile.ult + 17)
testfile.transform.min()

testfile.xy = testfile.dlc.reshape((-1, 11, 2)).copy()
testfile.both_xy = testfile.bothfilt.reshape((-1, 11, 2)).copy()
testfile.med_xy = testfile.medfilt.reshape((-1, 11, 2)).copy()
testfile.butter_xy = testfile.butter.reshape((-1, 11, 2)).copy()
testfile.col_names = ['vallecula_x', 'vallecula_y', 'tongueRoot1_x', 'tongueRoot1_y',
       'tongueRoot2_x', 'tongueRoot2_y', 'tongueBody1_x', 'tongueBody1_y',
       'tongueBody2_x', 'tongueBody2_y', 'tongueDorsum1_x', 'tongueDorsum1_y',
       'tongueDorsum2_x', 'tongueDorsum2_y', 'tongueBlade1_x',
       'tongueBlade1_y', 'tongueBlade2_x', 'tongueBlade2_y', 'tongueTip1_x',
       'tongueTip1_y', 'tongueTip2_x', 'tongueTip2_y']
```

```python
grid_dict = {
    'Butterworth Filter': testfile.butter,
    'Median Filter': testfile.medfilt,
    'Both Filters':testfile.bothfilt
}
```

```python
# show how each filter changes original signal
ppmm = 3.1625
plot_dir = 'speech2ult/figures/FilterExaples'
grid = ['Butterworth Filter', 'Median Filter', 'Both Filters']
fig, ax = plt.subplots(layout='constrained')
for col in range(len(testfile.col_names)):
# for col in range(1):
    xdata = np.arange(len(testfile.dlc[:, col])) / 60
    ax.set_title(f'Filtered vs Extracted DLC values for {testfile.col_names[col]}')
    ax.set_xlabel('Time (seconds)')
    ax.set_ylabel('Position (mm)')
    ax.set_xlim([0, xdata.max()])
    ax.plot(xdata, testfile.dlc[:, col]/ ppmm, label='Extracted DLC', color='tab:red', linewidth = 1)
    for k in range(3):
        ax.plot(xdata, grid_dict[grid[k]][:, col]/ppmm, label=grid[k], ls='--', linewidth = 1.5)
    ax.legend()
    plt.savefig(plot_dir + f"/{testfile.col_names[col]}.jpg", dpi=200)
    ax.clear()
```

```python
# figure for experiment 1
model_path = 'speech2ult/predictions/models/new_tests/'
arch = ['ffn', 'lstm']
models = ['model1', 'model2', 'model3']
model_list = [glob.glob(model_path + f'{a}/{m}/*.json')[0] for a in arch for m in models]
model_names = [f'{a.upper()}_{m.upper()}' for a in arch for m in models]
pred_list = []
```

```python
for path in model_list:
    model = load_model(path)

    if 'lstm' in path:
        pred_ult = model.predict(testfile.lstm, verbose=0)
    else:
        pred_ult = model.predict(testfile.aud_feat, verbose=0)

    pred_trans = testfile.ult_scalar.inverse_transform(pred_ult)
    pred_trans = transfrom(pred_trans.reshape(-1, 64, 128))
    pred_min = pred_trans.min()
    pred_trans = testfile.ult_scalar.inverse_transform(pred_ult)
    pred_trans = transfrom(pred_trans.reshape(-1, 64, 128) - pred_min)
    print(pred_min, pred_trans.min())
    pred_list.append(pred_trans)
```

```python
#exp2
model_path = 'speech2ult/predictions/models/dlc_tests/'
arch = ['ffn_dlc', 'ffn_dlc_both', 'ffn_dlc_butter', 'ffn_dlc_med']
models = ['model', 'model2']
model_list = [glob.glob(model_path + f'{a}/{m}/*.json')[0] for a in arch for m in models]
model_names = [f'{a.upper()}_{m.upper()}' for a in arch for m in models]
scaler_list = [glob.glob(model_path + f'{a}/data/PARAM*.pickle')[0] for a in arch for m in models]
pred_list = []
```

```python
model_names = ['Orig Model1', 'Orig Model2', 'Both Model1', 'Both Model2', 'Butter Model1', 'Butter Model2', 'Median Model1', 'Median Model2']
pred_xy = []
```

```python
for scal, path in zip(scaler_list, model_list):
    model = load_model(path)

    pred_dlc = model.predict(testfile.aud_feat, verbose=0)

    # get scaler
    with open(scal, 'rb') as file:
        dlc_params, aud_params = pickle.load(file)
        filt_scaler = dlc_params['scalar']

    pred_dlc = filt_scaler.inverse_transform(pred_dlc)
    pred_list.append(pred_dlc)
    pred_xy.append(pred_dlc.reshape((-1, 11, 2)))
```
